```python
import pandas as pd
import os

# Load the new dataset
file_path = r"C:\Users\afrin\OneDrive\Desktop\Air Quality Analytics & Pollution Forecasting\data\raw\delhi_ncr_aqi_dataset.csv"

df = pd.read_csv(file_path)
print("✅ Dataset Loaded!")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

# Basic Info
print("\nDelhi Data Count:", len(df[df['city'] == 'Delhi']))

# Save a copy for Delhi only (recommended)
delhi_df = df[df['city'] == 'Delhi'].copy()
os.makedirs("data/processed", exist_ok=True)
delhi_df.to_csv("data/processed/delhi_hourly_aqi.csv", index=False)
print("✅ Delhi data saved to data/processed/delhi_hourly_aqi.csv")
```
