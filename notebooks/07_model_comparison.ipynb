```python
import pandas as pd

comparison = pd.DataFrame({

'Model': [

'Random Forest',

'XGBoost',

'Prophet'

],


'R2': [

0.9999991417917898,

0.9999831318855286,

0.9265097662656283

],


'MAE': [

0.05034006983591996,

0.4436267018318176,

36.92731256753004

],


'RMSE': [

0.16555689582995564,

0.7344431744037265,

48.43738750255706

]

})


comparison
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(

comparison['Model'],

comparison['R2'],

color=['blue','green','deeppink']

)

plt.title("Model Comparison")


plt.ylabel("R² Score")


plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(

comparison['Model'],

comparison['RMSE'],

color=['orange','red','purple']

)


plt.ylabel("RMSE")

plt.title(

"RMSE Comparison"

)

plt.show()
```
