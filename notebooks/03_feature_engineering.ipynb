```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("data/processed/delhi_cleaned.csv")

df.head()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])

df = df.sort_values('datetime')

df.head()
```

```python
df['aqi_lag1'] = df['aqi'].shift(1)

df['aqi_lag3'] = df['aqi'].shift(3)

df['aqi_lag7'] = df['aqi'].shift(7)
```

```python
df['pm25_roll_mean'] = (

    df['pm25']

    .rolling(7)

    .mean()

)


df['pm25_roll_std'] = (

    df['pm25']

    .rolling(7)

    .std()

)



df['pm10_roll_mean'] = (

    df['pm10']

    .rolling(7)

    .mean()

)



df['pm10_roll_std'] = (

    df['pm10']

    .rolling(7)

    .std()

)
```

```python
print(df['season'].unique())
encoder = LabelEncoder()

df['season_encoded'] = encoder.fit_transform(df['season'])
```

```python
encoder2 = LabelEncoder()


df['aqi_category_encoded'] = (

encoder2.fit_transform(

df['aqi_category']

)

)
```

```python
df['pm_ratio'] = (

df['pm25']

/

df['pm10']

)



df['wind_humidity'] = (

df['wind_speed']

*

df['humidity']

)
```

```python
df=df.dropna()


print(df.shape)
```

```python
scaler = StandardScaler()


features=[


'pm25',

'pm10',

'no2',

'so2',

'co',

'o3',

'temperature',

'humidity',

'wind_speed',

'visibility',

'aqi_lag1',

'aqi_lag3',

'aqi_lag7',

'pm25_roll_mean',

'pm25_roll_std',

'pm10_roll_mean',

'pm10_roll_std',

'season_encoded',

'aqi_category_encoded',

'pm_ratio',

'wind_humidity'

]


df[features]=scaler.fit_transform(


df[features]

)
```

```python
df.to_csv(

"data/processed/delhi_featured.csv",

index=False

)


print(

"✅ Feature Engineering Complete"

)
```
